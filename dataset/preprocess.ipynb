```python
import os
import numpy as np
import pandas as pd
```

```python
def preprocess_df(df, fileName):
    print("\nOriginal - {}: {}".format(fileName, df.shape))
    df = df.replace("?", np.nan)
    df = df.dropna(how='any') 
    df.to_csv(os.path.join("preprocess", fileName), index=False)
    df = pd.read_csv(os.path.join("preprocess", fileName))
    df = df._get_numeric_data()
    df.to_csv(os.path.join("preprocess", fileName), index=False)
    print("Preprocess - {}: {}\n".format(fileName, df.shape))
    return df
```

```python
df = pd.read_excel(os.path.join("original", "Immunotherapy.xlsx"))
df = preprocess_df(df, "Immuno_Therapy.csv")
df.head()
```

```python
df = pd.read_csv(os.path.join("original", "breast-cancer-wisconsin.data"))
df = preprocess_df(df, "Breast_Cancer.csv")
df.head()
```

```python
cols = ["Class"] + ["Attribute"+str(i+1) for i in range(56)]
df = pd.read_csv(os.path.join("original", "lung-cancer.data"), names=cols)
cols = cols[1:] + cols[:1]
df = df[cols]
df = preprocess_df(df, "Lung_Cancer.csv")
df.head()
```

```python
cols = ["Attribute"+str(i+1) for i in range(24)] + ["Class"]
df = pd.read_csv(os.path.join("original", "german.data-numeric"), delim_whitespace=True, names=cols)
df = preprocess_df(df, "German_Credit.csv")
df.head()
```

```python
df = pd.read_csv(os.path.join("original", "setapProcess.csv"))
df['SE Process grade'] = df['SE Process grade'].map({'A': 1, 'F': 0})
df = preprocess_df(df, "SE_Process.csv")
df.head()
```

```python
fileName = "LSVT_voice_rehabilitation.xlsx"
df = pd.read_excel(os.path.join("original", fileName), sheet_name="Data")
dfLabel = pd.read_excel(os.path.join("original", fileName), sheet_name="Binary response", names=["Binary class"])
df["Binary class"] = dfLabel["Binary class"] 
df = preprocess_df(df, "Voice_Rehabilitation.csv")
df.head()
```
